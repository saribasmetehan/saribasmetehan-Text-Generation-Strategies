```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_name = "gpt2"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
model.generation_config.pad_token_id = model.generation_config.eos_token_id
```

```python
input = "As the president of the newly formed Turkish Republic, Atatürk initiated a rigorous program of political, economic, and cultural reforms with the ultimate aim of building a republican and secular nation-state."
```

```python
input_tokens = tokenizer(input, return_tensors="pt")
```

```python
input_tokens
```

```python
beam_outout = model.generate(
    **input_tokens,
    max_new_tokens = 50,
    num_beams = 7,
    early_stopping = True,
)
```

```python
print(tokenizer.decode(beam_outout[0], skip_special_tokens=True))
```

```python
beam_outout2 = model.generate(
    **input_tokens,
    max_new_tokens = 50,
    num_beams = 7,
    early_stopping = True,
    no_repeat_ngram_size = 2
)
```

```python
print(tokenizer.decode(beam_outout2[0], skip_special_tokens=True))
```

```python
beam_outout3 = model.generate(
    **input_tokens,
    max_new_tokens = 70,
    num_beams = 7,
    early_stopping = True,
    no_repeat_ngram_size = 2,
    num_return_sequences = 5
)
```

```python
for i, beam_outout3 in enumerate(beam_outout3):
    print(f"Output {i}:")
    print(tokenizer.decode(beam_outout3, skip_special_tokens=True))
```
